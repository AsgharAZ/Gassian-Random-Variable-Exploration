### [10 Marks] Task 2 : Expectation and Variance ###
In Task 1, you plotted the histogram for simulated throwing of a fair 6-sided die N times. In Task 2, you are required to compute the expectation and variance for each value of N and compare with the actual expectation and variance.  For this purpose, you will have to implement the following functions: compute_expectation, compute_variance.


```python
# Compute Expectation
#Syed Asghar Abbas Zaidi 07201
def compute_expectation (x: np.array, p: np.array ):
   expectation = 0;
   size = len(x);
   for i in range(size): #Adding all the Random Variables * theirProbability Mass Function to find the expected value
       expectation += x[i]*p[i]

   return expectation;

# Compute Variance
def compute_variance (x: np.array, p: np.array ):
    #Using the second formula to calculate variance E[X^2] - E([X])^2

    expected = compute_expectation (x,p); #Don't need to calculated expected to loop, we can just call the helper function to do it for us.
    expected_whole_squared = expected**2; #E([X])^2

    size = len(x);
    expected_X_squared = 0;
    for i in range(size):
      expected_X_squared += x[i]*x[i]*p[i] #calculating E[X^2] that's why we see x[i] multiplying with itself and then p[i]
    #variance = E[X^2]            -       E([X])^2
    variance = expected_X_squared - expected_whole_squared;
    return variance


```

```python
#Syed Asghar Abbas Zaidi 07201
def compute_expectation_variance (N: list):
    for n in N:
      print ("N = {val}".format(val = n ))
      # Simulate the rolling of a fair 6-sided die
      q = np.random.randint(low=1,high=7,size=n)
      x,p = np.unique(q,return_counts=True)
      p = p/n
      # Compute Expectation
      EX = compute_expectation(x,p)
      # Compute Variance
      Var = compute_variance(x,p)

      print ("Expectation : {ex:.2f} ,  Variance : {var:.2f}".format(ex=EX,var=Var))

compute_expectation_variance([100,1000,10000,100000])
```

In Task 1, you were introducted to the relationship between histogram and PMF of Random Variables. In Task 2, you were introduced to computing Expectation and Variance. The following code snippets show you how to simulate the generation of 1000 samples of the following types of random variables: Bernoulli, Binomial, Geometric, Poisson. The code also plots the histogram for the 1000 samples of each type of random variable.

```python
# Generate 1000 samples of a Bernoulli Random Variable
p = 0.5 # Probability of Success
n = 1   # Number of states (When n = 1, Binomial Random Variable becomes a Bernoulli Random Variable)
size  = 1000 # Number of trials
X = np.random.binomial(n,p,size=size)
plt.title("Bernoulli Random Variable (p=0.5)")
plt.hist(X,bins='auto')
plt.show()
```

```python
# Generate 1000 samples of a Binomial Random Variable
p = 0.5 # Probability of a single success
n = 10  # Number of states
size  = 1000 # Number of trials
X = np.random.binomial(n,p,size=size)
plt.title("Binomial Random Variable (n=10, p=0.5)")
plt.hist(X,bins="auto");
plt.show()
```

```python
# Generate 1000 samples of a Geometric Random variables
p = 0.5
size = 1000
plt.title("Geometric Random Variable (p = 0.5)")
X = np.random.geometric(p,size=size)
plt.hist(X,bins="auto")
plt.show()
```

```python
# Generate 1000 samples of a Poisson Random Variable
lambd = 1
X = np.random.poisson(lambd,size=1000)
plt.title("Poisson Random Variable (lambda = 1)")
plt.hist(X,bins="auto")
plt.show()
```

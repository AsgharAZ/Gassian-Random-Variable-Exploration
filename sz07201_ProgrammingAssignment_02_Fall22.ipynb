# [20 Points] Task 01 - Gaussian Random Variable

Gaussian random variables are the most common type of random variable. They are characterized by a mean $\mu$ and a standard deviation $\sigma$. The probability density function of a Gaussian random variable is given by
$$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $$
The standard normal distribution is the distribution of the Gaussian random variable with mean 0 and standard deviation 1 which is denoted by $ Z \sim \mathcal{N}(0,1)$. In this task, you are required to comment on the changes of the bell curve for different values of $\mu$ and $\sigma$.

## [10 points] Part A: Varying the Mean
Comment on the changes of the graph of the pdf when $\mu$ is varied and $\sigma = 1$ is kept constant.

```python
# DO NOT EDIT
# Plot pdf of the Gaussian distribution while varying mu and keeping sigma fixed
MU = [0, 1, 2, 3, 4, 5]
sigma = 1
plt.figure(figsize=(10,6))
for mu in MU:
    X = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    Y = stats.norm.pdf(X, mu, sigma)
    plt.plot(X, Y, label='mu = ' + str(mu))
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gaussian distribution')
plt.show()
```

Aside of this, changing the value of mu (mean) doesn't seem to affect the shap (thus standard deviation), height or any other property for that matter.

## [10 points] Part B: Varying the Standard Deviation
Plot PDF of the Gaussian random variable while keeping the mean $\mu$ fixed at 0 and and varying $\sigma$ to folliwig values: 0.2, 0.5, 1.0, 2.0.

Comment on the changes in the PDF shape when $\sigma$ is varied and $\mu = 0$ is kept constant.

```python
# Syed Asghar Abbas Zaidi
# Student ID: sz07201

MU = 0 # mean will still remain as constant
sigma = [0.2, 0.5, 1.0, 2.0] # Standard Deviation will be varied.
plt.figure(figsize=(10,6))

for sd in sigma: # Iterating over all the values of standard deviation
    X = np.linspace(MU - 3*sd, MU + 3*sd, 100)
    Y = stats.norm.pdf(X, MU, sd)
    plt.plot(X, Y, label='Standard Deviation = ' + str(sd))

plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gaussian distribution')
plt.show()
```

# [30 points] Task 02 - Central Limit Theorem

Suppose $X_1,X_2, \cdots, X_n$ are independent random variables with the same underlying distribution. In this case, we say that the $X_i$ are independent and identically distributed or i.i.d. In particular, the $X_i$, all have the same mean $\mu$ and standard deviation $\sigma$.

Let $S_n$ be the sum of $n$ i.i.d random variables:
$$ S_n = X_1 + X_2 + \cdots + X_n $$


## [10 points] Part A: Uniform Random Variable
In this part, we consider $X_i$s to be continuous uniform random variables. The following code generates 10,000 samples of $X_1$ random variable and plots an approximate PDF of $X_1$ (based on its 10,000 samples) using the displot() function of Seaborn library.

Modify the given code to generate 10,000 samples of $X_2,X_3, \cdots, X_n$, and plot the approximate PDF of $S_n$ using the displot() function of Seaborn library for the following values of n: 1,2,3,5,10,50,100.

Comment on the evolution of PDF shape of $S_n$ as n increases.

```python
#Syed Asghar Abbas Zaidi 07201
sample_size = 10000
i = [1,2,3,5,10,50,100]
for n in i:
  df = pd.DataFrame()
  x = np.random.uniform(0, n, sample_size)
  col = f'n = {n}'
  df[col] = x

  sns.displot(data=df[col], kind="kde")
```

**Syed Asghar Abbas Zaidi**

*Comments:*

As n increases, we can observe our graph's spread increasing and it's PDF or Density decreasing. Take for example, n = 1, its' width is pretty small however its height is equally far more larger. However, on a greater value of n, like n = 100, the width is very large but the height is noticeably very small. This property can again be attributed to the fact that the area underneath the graph must always equal to "1" if it wants to be a valid PDF graph.


## [10 points] Part B: Exponential Random Variable
In this part, we consider $X_i$s to be exponential random variables. The following code generates 10,000 samples of $X_1$ random variable and plots an approximate PDF of $X_1$ (based on its 10,000 samples) using the displot() function of Seaborn library.

Modify the given code to generate 10,000 samples of $X_2,X_3, \cdots, X_n$, and plot the approximate PDF of $S_n$ using the displot() function of Seaborn library for the following values of n: 1,2,3,5,10,50,100.

Comment on the evolution of PDF shape of $S_n$ as n increases.

```python
# Syed Asghar Abbas Zaidi 07201
sample_size = 10000
rate = 0.25 #Rate of exponential random variables
i = [1,2,3,5,10,50,100]
for n in i:
  df = pd.DataFrame()
  x = np.random.exponential((1/(rate * n)),sample_size)
  col = f'n = {n}'
  df[col] = x
  sns.displot(data=df[col], kind="kde")
```

**Syed Asghar Abbas Zaidi 07201**

*Comments:*

Increasing the value of n, we observe that pmf extends and is distributed to higher values, thus the width of the graph decreases and height increases. Vice versa is again true, in smaller values of n, the width increases, graph becomes more spread out and height decreases. This relationship exists to make sure that the total pmf always remains equal to 1 that's why relative pdfs get scaled accordingly. Aside of that, they all share the same shape despite big changes in size is cause all of them have identical rates.
